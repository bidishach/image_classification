```python
import tensorflow as tf
from tensorflow.keras.layers.experimental import preprocessing
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Dropout
from tensorflow.keras.layers import BatchNormalization, Flatten, Rescaling
from tensorflow.keras.utils import image_dataset_from_directory
import matplotlib.pyplot as plt
```

## Step 1

-    Took input coloured Images of Cats and Dogs. 
-     Now, using the tensorflows utility method `Image Dataset from Directory`, we will make our source dataset of type `tf.data.dataset` from the input images. 

Few basic features of the below dataset are:
-  Image size is ( 299 * 299 * 3), since the default color mode is 'rgb'
-  Images will be stored as tensor and batch size is 4.
-  Train and validation split is 80:20

### Train Dataset

```python
traindata = image_dataset_from_directory("dataset/",
                                         image_size=(299,299),
                                         batch_size=4,
                                         seed=10,
                                         validation_split=0.2,
                                         subset="training")
```

### Validation Dataset

```python
validationdata = image_dataset_from_directory("dataset/",
                                         image_size=(299,299),
                                         batch_size=4,
                                         seed=10,
                                         validation_split=0.2,
                                         subset="validation") 
```

```python
for image,label in traindata:
    print(image.shape, label.shape)
```

```python
for image,label in validationdata:
    print(image.shape, label.shape)
```

This is going to be a binary image clasification.

The class names are: 

```python
class_names = traindata.class_names
class_names
```

### Displaying random samples from training dataset

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(6, 6))
for images, labels in traindata.take(2):
    labels = np.array(labels).astype(int)
    for i in range(4):
        ax = plt.subplot(2, 2, i + 1)    
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off");
```

Now lets create our CNN Model using two convolution layers

```python
model= Sequential()

# Input layer along with scaling
model.add(Rescaling(scale=1./255,input_shape= ( 299, 299, 3)))

# Conv layer 1, 
# number of filters is 64,
# filter/Kernel size is 3*3 and
#padding is default padding "same"
model.add(Conv2D(64,(3,3),activation="relu"))

# MaxPooling layer 1
model.add(MaxPooling2D())

# Conv layer 2
# number of filters is 32,
# filter/Kernel size is 3*3 and
#padding is default padding "same"
model.add(Conv2D(32,(3,3),activation="relu"))

# MaxPooling layer 2
model.add(MaxPooling2D())

#Flatten layer
model.add(Flatten())

#ANN i.e. Dense layer 1
model.add(Dense(128,activation="relu"))

# Output Layer 
model.add(Dense(1,activation="sigmoid"))
```

Now it is time to provide the attributes that would help compile the method during training

```python
model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=["accuracy"])
```

Train the custom CNN Model

```python
history = model.fit(traindata, epochs = 10, validation_data = validationdata)
```

### Chek if the model is overfitting

Plot: 
- Training Accuracy vs Validation Accuracy 
- Training Loss vs Validation Loss

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend();

plt.figure()
plt.plot(epochs, loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label = 'Validation loss')
plt.title('Training and validation loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend()
plt.show();
```

From the above plot it is obvious that our model is overfitting since the training accuracy is around 0.97 but validation accuracy is still at 0.25. 

So, now we will use some **regularization methods** like
- Dropout and Batch Normalization
- augmentation
- early stopping

In order to improve our existing model

```python
model= Sequential()

# Input layer along with scaling
model.add(Rescaling(scale=1./255,input_shape= ( 299, 299, 3)))

# Now lets add some data augmentation steps    
# 1. `translation`
model.add(preprocessing.RandomTranslation(0.1,0.2)),
# 2. `flip` and 
model.add(preprocessing.RandomFlip('horizontal')),
# 3. `rotation transformations`.    
model.add(preprocessing.RandomRotation(0.1))

#     Layer - 1    
model.add(Conv2D(filters=128,kernel_size=(3,3),padding="same", activation="relu"))
model.add(BatchNormalization())
model.add(MaxPooling2D(strides=(2,2)))
#     Layer - 2
model.add(Conv2D(filters=64, kernel_size=(3,3), padding="same", activation="relu"))
model.add(BatchNormalization())
model.add(MaxPooling2D(strides=(2,2)))
#     Layer - 3    
model.add(Conv2D(filters=32, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPooling2D(strides=(2,2)))
#     Dense Layer - 1 
model.add(Flatten())
#model.add(Dense(64, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=["accuracy"])
```

```python
earlyStopping = tf.keras.callbacks.EarlyStopping(
                                                 monitor='val_loss', 
                                                 min_delta=0.0001, 
                                                 patience=5, 
                                                 restore_best_weights=True
                                                )
```

```python
history = model.fit(traindata, epochs = 20, validation_data = validationdata, callbacks= [earlyStopping])
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend();
```

### After adding regularisation methods the model improved a lot since overfitting has reduced sufficiently.
